## 🎯 Essential Tool: The `input()` Function

### What is `input()`?

The `input()` function allows your program to **ask the user for information**:
- **Syntax**: `variable = input("Your question here")`
- The user types their answer
- The answer is **always stored as text (string)**
- To use numbers, you must convert: `int(input(...))` or `float(input(...))`

### Examples:

```python
# Get a name
name = input("What is your name? ")
print(f"Hello, {name}!")

# Get a number (must convert to int)
age = int(input("How old are you? "))
print(f"You are {age} years old.")

# Get a decimal number
height = float(input("What is your height in meters? "))
print(f"Your height is {height}m")
```

### Key Rules:
1. `input()` **always returns a string**
2. Use `int()` to convert to whole numbers
3. Use `float()` to convert to decimal numbers
4. The text inside `input("...")` appears as a **prompt** to the user

## 📌 Concept 1: The `if` Statement

### What is an `if` Statement?

An `if` statement allows your code to make a **decision**:
- **Syntax**: `if condition:`
- The code under `if` runs **only if** the condition is `True`
- Conditions use **comparison operators**: `==`, `!=`, `>`, `<`, `>=`, `<=`

### Example:
```python
age = 18
if age >= 18:
    print("You are an adult!")
```

**Output**: `You are an adult!`

### How it Works:
1. Check the condition: `age >= 18` → `18 >= 18` → **True**
2. Since it's `True`, execute the indented code
3. Print the message

````python
# Example 1: Simple if statement
score = int(input("Enter your test score: "))

if score >= 80:
    print("Great job! You passed!")
    print(f"Your score is {score}")

# Example 2: Check if a number is positive
number = int(input("Enter a number: "))

if number > 0:
    print(f"{number} is a positive number")

# Example 3: Check if something is in a list
fruits = ["apple", "banana", "orange"]
fruit_name = input("Are we looking for (apple/banana/orange)? ")

if fruit_name in fruits:
    print(f"We have {fruit_name}!")
````

## 📌 Concept 2: The `if-else` Statement

### What is `if-else`?

Sometimes you want to do **something different** when the condition is false:
- **Syntax**:
  ```python
  if condition:
      # runs if True
  else:
      # runs if False
  ```

### Example:
```python
age = 16
if age >= 18:
    print("You can vote!")
else:
    print("You're too young to vote.")
```

**Output**: `You're too young to vote.`

### How it Works:
1. Check: `age >= 18` → `16 >= 18` → **False**
2. Since it's `False`, skip the `if` block
3. Execute the `else` block
4. Print the message

````python
# Example 1: Checking a password
correct_password = "secret123"
user_input = input("Enter password: ")

if user_input == correct_password:
    print("Access granted!")
else:
    print("Wrong password!")

# Example 2: Even or Odd
num = int(input("Enter a number: "))

if num % 2 == 0:
    print(f"{num} is even")
else:
    print(f"{num} is odd")

# Example 3: Weather decision
temperature = int(input("Enter temperature (°C): "))

if temperature > 20:
    print("It's warm! Go outside.")
else:
    print("It's cold. Stay inside.")
````

## 📌 Concept 3: The `if-elif-else` Statement

### What is `if-elif-else`?

When you have **multiple conditions** to check:
- **Syntax**:
  ```python
  if condition1:
      # runs if condition1 is True
  elif condition2:
      # runs if condition1 is False AND condition2 is True
  else:
      # runs if all conditions are False
  ```

### Key Points:
- **`elif`** = "else if" - check another condition
- You can have **multiple `elif` blocks**
- Only **one block executes** (the first true one)

### Example:
```python
score = 75
if score >= 90:
    print("Grade: A")
elif score >= 80:
    print("Grade: B")
elif score >= 70:
    print("Grade: C")
else:
    print("Grade: F")
```

**Output**: `Grade: C`

````python
# Example 1: Traffic light
light = input("Enter light color (red/yellow/green): ")

if light == "red":
    print("STOP!")
elif light == "yellow":
    print("Prepare to stop")
elif light == "green":
    print("GO!")
else:
    print("Invalid light")

# Example 2: Size categorization
size = int(input("Enter a size (0-200): "))

if size < 30:
    print("Small")
elif size < 50:
    print("Medium")
elif size < 100:
    print("Large")
else:
    print("Extra Large")

# Example 3: Age groups
age = int(input("Enter your age: "))

if age < 13:
    print("Child")
elif age < 18:
    print("Teen")
elif age < 65:
    print("Adult")
else:
    print("Senior")
````

## 📌 Concept 4: Nested Conditionals & Logical Operators

### What are Nested Conditionals?

**Nested** means putting `if` statements **inside** other `if` statements:

```python
if age >= 18:
    if license:
        print("You can drive!")
```

### Logical Operators for Complex Conditions

Instead of nesting, you can use **logical operators**:
- **`and`**: ALL conditions must be True
- **`or`**: AT LEAST ONE condition must be True
- **`not`**: Reverses True/False

### Examples:

```python
# Using 'and'
age = 20
license = True
if age >= 18 and license:
    print("You can drive!")

# Using 'or'
is_student = True
is_senior = False
if is_student or is_senior:
    print("Discount available!")

# Using 'not'
raining = False
if not raining:
    print("Go outside!")
```

````python
# Example 1: Nested conditionals
balance = int(input("Enter your account balance ($): "))
amount = int(input("How much do you want to withdraw ($): "))

if amount > 0:
    if amount <= balance:
        print(f"Withdrawal approved! New balance: ${balance - amount}")
    else:
        print("Insufficient funds!")
else:
    print("Invalid amount!")

# Example 2: Using 'and'
temperature = int(input("Enter temperature (°C): "))
humidity = int(input("Enter humidity (%): "))

if temperature > 20 and humidity < 70:
    print("Perfect weather for outdoor activities!")

# Example 3: Using 'or'
print("Are you eligible for a discount?")
has_coupon = input("Do you have a coupon? (yes/no): ") == "yes"
is_member = input("Are you a member? (yes/no): ") == "yes"

if has_coupon or is_member:
    print("You get a discount!")

# Example 4: Using 'not'
is_broken = input("Is the machine broken? (yes/no): ") == "yes"

if not is_broken:
    print("Machine is working correctly.")
````

````python
# Capstone Project: Complete your solution here

score = int(input("Enter student score (0-100): "))
attendance = int(input("Enter attendance percentage (0-100): "))

# Step 1: Determine grade letter
if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"

# Step 2: Check attendance
good_grade = score >= 80
good_attendance = attendance >= 80

# Step 3: Provide feedback using nested conditionals
if good_grade and good_attendance:
    feedback = "Excellent work! Keep it up!"
elif good_grade and not good_attendance:
    feedback = "Good grade, but improve attendance."
elif not good_grade and good_attendance:
    feedback = "You attend well but need to study more."
else:
    feedback = "Both grades and attendance need improvement."

# Step 4: Display results
print(f"\n{'='*40}")
print(f"Score: {score}, Attendance: {attendance}%")
print(f"Grade: {grade}")
print(f"Feedback: {feedback}")
print(f"{'='*40}")
````

## 📋 Key Takeaways

| Concept | Syntax | Use Case |
|---------|--------|----------|
| **input()** | `variable = input("prompt")` | Get data from user (always a string) |
| **int()** | `int(input(...))` | Convert string to integer |
| **float()** | `float(input(...))` | Convert string to decimal number |
| **if** | `if condition:` | Execute code only when condition is True |
| **if-else** | `if condition: ... else: ...` | Choose between two options |
| **if-elif-else** | `if c1: ... elif c2: ... else: ...` | Multiple conditions |
| **and** | `condition1 and condition2` | ALL must be True |
| **or** | `condition1 or condition2` | AT LEAST ONE must be True |
| **not** | `not condition` | Reverse True/False |

## 🎓 What You Learned

✅ Use `input()` function to get user data  
✅ Convert user input to numbers using `int()` and `float()`  
✅ Use `if` statements for single conditions  

✅ Use `if-else` for two-way decisions  ✅ Create interactive programs that respond to user input

✅ Use `if-elif-else` for multiple conditions  ✅ Build complex decision systems  
✅ Combine conditions with `and`, `or`, `not`
