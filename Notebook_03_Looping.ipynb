## What is a Loop?

A **loop** is a block of code that **repeats multiple times**. Instead of writing the same code over and over, you can use a loop to:
- **Repeat code automatically** without rewriting it
- **Process multiple items** in a list or collection
- **Run code until a condition changes** (while loops)
- **Reduce code duplication** and make programs cleaner

This notebook focuses on **two main types of loops**: `for` loops and `while` loops.

## ğŸ¯ Essential Tool: The `range()` Function

### What is `range()`?

The `range()` function **generates a sequence of numbers** that you can loop through:
- **Syntax**: `range(start, stop, step)` or `range(stop)`
- Used with `for` loops to repeat code a specific number of times
- Creates numbers **starting at 0** by default
- Stops **before the final number**

### Common Examples:

```python
# Count from 0 to 4 (stops before 5)
range(5)  # produces: 0, 1, 2, 3, 4

# Count from 2 to 7 (starts at 2, stops before 8)
range(2, 8)  # produces: 2, 3, 4, 5, 6, 7

# Count by 2s: 0, 2, 4, 6
range(0, 7, 2)  # produces: 0, 2, 4, 6

# Count backward (negative step)
range(5, 0, -1)  # produces: 5, 4, 3, 2, 1
```

### How to Use with Loops:

```python
# Print numbers 1 to 5
for i in range(1, 6):
    print(i)

# Print "Hello" 3 times
for i in range(3):
    print("Hello")
```
