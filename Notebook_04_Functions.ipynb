## 1. Basic Function Definition

### Simple Function
A function is a reusable block of code that performs a specific task

```python
# Simple function with no parameters
def greet():
    print("Hello! Welcome to Python.")

# Call the function
greet()
greet()
```

## 2. Functions with Parameters

### Function with One Parameter

```python
# Function with one parameter
def greet_person(name):
    print(f"Hello, {name}!")

greet_person("Alice")
greet_person("Bob")
```

### Function with Multiple Parameters

```python
# Function with multiple parameters
def calculate_area(length, width):
    area = length * width
    print(f"Rectangle: {length} x {width} = {area}")

calculate_area(5, 10)
calculate_area(3, 8)
```

## 3. Return Values

### Returning a Single Value

```python
# Function that returns a value
def add_numbers(a, b):
    result = a + b
    return result

sum1 = add_numbers(5, 3)
sum2 = add_numbers(10, 20)

print(f"Sum 1: {sum1}")
print(f"Sum 2: {sum2}")
```

### Returning Multiple Values

```python
# Function returning multiple values as a tuple
def get_min_max(numbers):
    min_val = min(numbers)
    max_val = max(numbers)
    return min_val, max_val

numbers = [45, 23, 89, 12, 67]
minimum, maximum = get_min_max(numbers)

print(f"Numbers: {numbers}")
print(f"Min: {minimum}, Max: {maximum}")
```

## 4. Default Parameters

### Using Default Values

```python
# Function with default parameters
def power(base, exponent=2):
    return base ** exponent

# Called with both parameters
print(f"5^2 = {power(5, 2)}")

# Called with default exponent
print(f"5^2 (default) = {power(5)}")
print(f"3^3 = {power(3, 3)}")
```

## 5. Variable Scope

### Local vs Global Variables

```python
# Global variable
global_var = "I'm global"

def my_function():
    # Local variable
    local_var = "I'm local"
    print(f"Inside function: {local_var}")
    print(f"Inside function: {global_var}")

my_function()

print(f"Outside function: {global_var}")
# print(local_var)  # This would cause an error!
```

### Modifying Global Variables
Use the `global` keyword to modify global variables inside a function

```python
# Working with global variables
counter = 0

def increment():
    global counter
    counter += 1

print(f"Initial counter: {counter}")
increment()
print(f"After increment: {counter}")
increment()
print(f"After second increment: {counter}")
```

## 6. Documentation and Type Hints

### Docstrings
Describe what your function does

```python
# Function with docstring
def celsius_to_fahrenheit(celsius):
    """
    Convert temperature from Celsius to Fahrenheit.
    
    Parameters:
    celsius (float): Temperature in Celsius
    
    Returns:
    float: Temperature in Fahrenheit
    """
    fahrenheit = (celsius * 9/5) + 32
    return fahrenheit

# Access the docstring
print(celsius_to_fahrenheit.__doc__)
print(f"0Â°C = {celsius_to_fahrenheit(0)}Â°F")
```

## 7. Function Best Practices

### Example: Well-Written Function

```python
# Good function example
def calculate_average(numbers):
    """
    Calculate the average of a list of numbers.
    
    Args:
        numbers (list): A list of numeric values
    
    Returns:
        float: The average of the numbers
    """
    if len(numbers) == 0:
        return 0
    
    total = sum(numbers)
    average = total / len(numbers)
    return average

grades = [85, 90, 78, 92, 88]
avg = calculate_average(grades)
print(f"Average grade: {avg:.2f}")
```
