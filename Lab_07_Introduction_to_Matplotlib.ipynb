## ما هي مكتبة Matplotlib؟

</div>
</details>

Matplotlib is a powerful data visualization library for Python. It provides a wide range of tools and functions for creating various types of plots, charts, and graphs, making it an essential tool for data analysts, scientists, and researchers.

With Matplotlib, you can create line plots, scatter plots, bar plots, histograms, pie charts, and many other types of visualizations. It offers a high level of customization, allowing you to control every aspect of your plots, including colors, markers, line styles, labels, titles, and legends.

Matplotlib is designed to be easy to use and integrates well with other scientific libraries in the Python ecosystem, such as NumPy and Pandas. It provides a simple and intuitive interface, making it accessible to both beginners and experienced users.

One of the key features of Matplotlib is its ability to create publication-quality plots. You can save your plots in various file formats, including PNG, PDF, SVG, and more. This makes it easy to incorporate your visualizations into reports, presentations, and scientific publications.

Matplotlib is an open-source library and has a large and active community of users. This means that you can find plenty of resources, tutorials, and examples online to help you learn and master the library. Additionally, Matplotlib has extensive documentation that covers all its features in detail.

John Hunter was the creator of Matplotlib. He developed Matplotlib in 2002 as a tool for his neuroscience research. Hunter's goal was to create a library that would allow him and other researchers to create high-quality visualizations quickly and easily.

Under Hunter's leadership, Matplotlib grew in popularity and became an essential tool in the Python data science ecosystem. He actively maintained and contributed to the library until his death in 2012 at an age of 44. Hunter was highly regarded for his technical expertise, dedication to scientific visualization, and his support of the open-source community.

Matplotlib allow us create various types of 2D charts and visualizations. The different types of charts that Matplotlib can create includes:

1. **_Line Plot_**:
A line plot is a simple chart that displays data points connected by straight lines. It is useful for visualizing trends and changes over time or a continuous variable.   

2. **_Scatter Plot_**:
A scatter plot represents individual data points as markers on a two-dimensional plane. It is commonly used to visualize the relationship or correlation between two variables.   

3. **_Bar Chart_**:
A bar chart uses rectangular bars of varying lengths to represent data. It is effective for comparing categorical or discrete data across different categories or groups.   

4. **_Histogram_**:
A histogram displays the distribution of a continuous variable by dividing it into bins and plotting the frequency or count of data points within each bin. It is useful for understanding the underlying distribution of data.   

5. **_Pie Chart_**:
A pie chart represents data as slices of a pie, where each slice corresponds to a different category or proportion of a whole. It is suitable for displaying relative proportions or percentages.

6. **_Area Plot_**:
An area plot is similar to a line plot, but the area beneath the line is filled with color, creating a visually stacked representation. It is useful for visualizing cumulative data or multiple variables over time.   

7. **_Box Plot_**:
A box plot, also known as a box-and-whisker plot, provides a summary of the distribution of a continuous variable. It displays the median, quartiles, and outliers of the data, making it useful for identifying skewness, variability, and outliers.   

8. **_Heatmap_**:
A heatmap uses colors to represent the intensity or magnitude of values in a two-dimensional matrix. It is commonly used to visualize correlations, patterns, or distributions in large datasets.   

9. **_Violin Plot_**:
A violin plot combines a box plot and a kernel density estimation to display the distribution of a variable. It provides more information about the density and shape of the data compared to a box plot alone.   

10. **_Polar Plot_**:
A polar plot represents data on a polar coordinate system, where the angle and distance from the origin represent different variables. It is useful for visualizing cyclic or directional data.   
In this introduction we will focus on the most commonly used ones.

Subplotting refer to the process of creating multiple plots within a single figure. In matplotlib, we can acheive this by using the function `plt.subplots()` which creates multiple plots in array form. This function takes two arguments:   

* **_nrows_**: specify how many rows of plots are required.The default is 1   
* **_ncols_**: specify how many columns of plots are required. The default is 1      

The function return a tuple of two objects. The first one is the figure object that contains all the plots and the second one is an array of created subplots which are referred to as axes. To explain the subplotting process in matplotlib let us create 4 plots (2 rows and 2 columns) in one figure as it is shown in the example below:

* **_nrows_**: تحديد عدد صفوف قطع الأراضي المطلوبة. الإعداد الافتراضي هو 1   
* **_ncols_**: تحديد عدد أعمدة قطع الأرض المطلوبة. الافتراضي هو 1

تعيد الدالة زوجًا من الكائنات. الأول هو كائن الشكل الذي يحتوي على جميع المخططات والثاني عبارة عن مجموعة من المخططات الفرعية التي تم إنشاؤها والتي يشار إليها بالمحاور. لشرح عملية الرسم الفرعي في Matplotlib دعونا ننشئ 4 مخططات (صفّان وعمودان) ضمن شكل واحد كما هو موضح في المثال أدناه:

Matplotlib allows us to save our plots as image files using the function `plt.savefig()`. This function takes one required argument (`filename`), which is the file name and path where you want to save the plot. The file extension in filename determines the format of the saved image. Let's take a line plot as an example and try to save it as a PNG file.

يتيح لنا Matplotlib حفظ مخططاتنا كملفات صور باستخدام الوظيفة `plt.savefig()`. تأخذ هذه الوظيفة وسيطة واحدة مطلوبة (`filename`)، وهي اسم الملف والمسار الذي تريد حفظ المخطط فيه. يحدد امتداد الملف في اسم الملف تنسيق الصورة المحفوظة. لنأخذ مخططًا خطيًا كمثال ونحاول حفظه كملف PNG.

</div>

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.title('My Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid()

# Save the plot as a PNG file
plt.savefig('line_plot.png')

# Display the plot
plt.show()
```

After running the code above, we will find a file named line_plot.png in the same directory that we are working in. We can also specify a different file path or format (e.g., PDF, JPEG, SVG) by modifying the argument passed to `plt.savefig()` function.

The `plt.savefig()` function provides also several optional parameters to customize the saved image. Some commonly used parameters include:

* **_dpi_**: Sets the resolution (dots per inch) of the saved image. Higher values result in higher quality but larger file sizes.   
* **_format_**: Specifies the file format for the saved image. Matplotlib automatically detects the format based on the file extension, but you can explicitly specify it if needed.   
* **_transparent_**: Controls whether the saved image has a transparent background. Set it to True for a transparent background.   

Here's an example that demonstrates some of these parameters:

<details>
<summary style="cursor: pointer; color: #667EEA; font-weight: bold; font-size: 14px; font-family: 'Amiri', serif;">🌍 Arabic Translation / الترجمة العربية</summary>

توفر وظيفة `plt.savefig()` أيضًا العديد من المعلمات الاختيارية لتخصيص الصورة المحفوظة. تتضمن بعض المعلمات شائعة الاستخدام ما يلي:

* **_dpi_**: ضبط الدقة (نقطة في البوصة) للصورة المحفوظة. تؤدي القيم الأعلى إلى جودة أعلى ولكن أحجام ملفات أكبر.   
* **_format_**: يحدد تنسيق الملف للصورة المحفوظة. يكتشف Matplotlib تلقائيًا التنسيق بناءً على امتداد الملف، ولكن يمكنك تحديده بشكل صريح إذا لزم الأمر.   
* **_transparent_**: يتحكم في ما إذا كانت الصورة المحفوظة لها خلفية شفافة. اضبطه على True للحصول على خلفية شفافة.

فيما يلي مثال يوضح بعض هذه المعلمات:

</div>

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.title('My Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid()

# Save the plot as a PNG file
plt.savefig('plot.png', dpi=300, format='png', transparent=True)

# Display the plot
plt.show()
```

In the example above, the plot is saved as a PNG image file with a resolution of 300 DPI, and a transparent background.
