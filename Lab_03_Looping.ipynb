- **Repeat code automatically** without rewriting it
- **Process multiple items** in a list or collection

- **Syntax**: `range(start, stop, step)` or `range(stop)`
- Used with `for` loops to repeat code a specific number of times

## 🎯 أداة أساسية: دالة `range()`

### ما هي `range()`؟

دالة `range()` **تولد سلسلة من الأرقام** يمكنك المرور عليها:
- **الصيغة**: `range(start, stop, step)` أو `range(stop)`
- تُستخدم مع حلقات `for` لتكرار الكود عدداً محدداً من المرات
- تبدأ الأرقام **من 0** افتراضياً
- تتوقف **قبل الرقم الأخير**

### أمثلة شائعة:

```python
# طباعة "مرحباً" 3 مرات
for i in range(3):
    print("مرحباً")

# طباعة الأرقام من 1 إلى 5
for i in range(1, 6):
    print(i)
```

### كيفية الاستخدام مع الحلقات:

```python
# العد العكسي (خطوة سلبية)
for num in range(5, 0, -1):
    print(num)

# العد بمقدار 2: 0, 2, 4, 6
for num in range(0, 7, 2):
    print(num, end=" ")

# العد من 2 إلى 7 (يبدأ من 2، يتوقف قبل 8)
for num in range(2, 8):
    print(num)

# العد من 0 إلى 4 (يتوقف قبل 5)
for num in range(5):
    print(num)
```

## 📌 Concept 1: The `for` Loop with `range()`

### What is a `for` Loop?

A `for` loop repeats code **a specific number of times**:
- **Syntax**:
```python
for variable in range(n):
    # code to repeat
```
- The `variable` takes each value from the range
- The code **indented** under the loop repeats each time
- Very useful when you know **how many times** to repeat

### Example:
```python
for i in range(3):
    print("Welcome!")
```

**Output**:
```
Welcome!
Welcome!
Welcome!
```

### How it Works:
1. First iteration: i=0, print "Welcome!"
2. Second iteration: i=1, print "Welcome!"
3. Third iteration: i=2, print "Welcome!"
4. Loop ends

## 📌 Concept 3: The `while` Loop

### What is a `while` Loop?

A `while` loop repeats code **while a condition is true**:
- **Syntax**:
```python
while condition:
    # code to repeat
```
- Useful when you **don't know how many times** to repeat
- Stops when the condition becomes False
- **Must change** something in the loop to eventually stop

### Example:
```python
count = 0
while count < 3:
    print(f"Count: {count}")
    count = count + 1
```

**Output**:
```
Count: 0
Count: 1
Count: 2
```

### Key Difference from `for`:
- `for`: Repeats a **specific number** of times
- `while`: Repeats **until a condition changes**

## 📌 المفهوم 3: حلقة `while`

### ما هي حلقة `while`؟

- **الصيغة**:
```python
while condition:
    # الكود المراد تكراره
```
- مفيدة عندما **لا تعرف كم مرة** تريد التكرار
- توقف عندما يصبح الشرط **خاطئاً**
- **يجب تغيير** شيء في الحلقة للتوقف في النهاية

### الفرق الرئيسي:

## 📌 Concept 4: `break` and `continue`

### What are `break` and `continue`?

These keywords control the **flow** of loops:
- **`break`**: Exits the loop **immediately**
- **`continue`**: Skips the **current iteration** and continues to the next

### Examples:

````python
# Using break - stop when finding a value
for i in range(10):
    if i == 5:
        break  # Exit loop
    print(i)
````

````python
# Using continue - skip even numbers
for i in range(5):
    if i % 2 == 0:
        continue  # Skip this iteration
    print(i)  # Only prints odd numbers: 1, 3
````

## 📌 Concept 5: Nested Loops

### What is a Nested Loop?

A **nested loop** is a loop **inside another loop**:
- The **outer loop** runs first
- For **each iteration** of the outer loop, the **inner loop** runs completely
- Very useful for working with **2D data**, **tables**, or **patterns**

### Syntax:
```python
for outer_var in range(outer_count):
    for inner_var in range(inner_count):
        # This code runs outer_count × inner_count times
```

### Example 1: Multiplication Table
```python
for i in range(1, 4):
    for j in range(1, 4):
        print(f"{i} × {j} = {i*j}")
```

**Output**:
```
1 × 1 = 1
1 × 2 = 2
1 × 3 = 3
2 × 1 = 2
2 × 2 = 4
2 × 3 = 6
3 × 1 = 3
3 × 2 = 6
3 × 3 = 9
```

### Example 2: Pattern Printing
```python
for row in range(3):
    for col in range(5):
        print("*", end=" ")
    print()  # New line after each row
```

**Output**:
```
* * * * *
* * * * *
* * * * *
```

### How It Works:
1. Outer loop starts: `row = 0`
2. Inner loop runs completely: prints 5 stars
3. New line is printed
4. Outer loop continues: `row = 1`
5. Inner loop runs again: prints 5 stars
6. This repeats until outer loop finishes

## 📌 المفهوم 5: الحلقات المتداخلة

### ما هي الحلقة المتداخلة؟

**الحلقة المتداخلة** هي حلقة **داخل حلقة أخرى**:
- **الحلقة الخارجية** تعمل أولاً
- في **كل تكرار** من الحلقة الخارجية، **الحلقة الداخلية** تعمل بالكامل
- مفيدة جداً للعمل مع **البيانات ثنائية الأبعاد**، **الجداول**، أو **الأنماط**

### الصيغة:
```python
for outer_var in range(outer_count):
    for inner_var in range(inner_count):
        # هذا الكود يعمل outer_count × inner_count مرة
```

### مثال 1: جدول الضرب
```python
for i in range(1, 4):
    for j in range(1, 4):
        print(f"{i} × {j} = {i*j}")

for row in range(3):

    print()  # سطر جديد بعد كل صف
```

**الإخراج**:
```

* * * * *
* * * * *
```

### كيف تعمل:
1. الحلقة الخارجية تبدأ: `row = 0`
2. الحلقة الداخلية تعمل بالكامل: تطبع 5 نجوم
3. طباعة سطر جديد
4. الحلقة الخارجية تستمر: `row = 1`
5. الحلقة الداخلية تعمل مرة أخرى: تطبع 5 نجوم
6. هذا يتكرر حتى تنتهي الحلقة الخارجية

````python
# Capstone Project: Complete your solution here

print("Multiplication Table Generator")
print("=" * 30)

# Outer while loop: Keep asking until user says no
while True:
    # Get input from user
    num = int(input("\nEnter a number: "))
    limit = int(input("Multiply up to: "))

    # Inner for loop: Generate multiplication table
    print(f"\nTable of {num}:")
    for i in range(1, limit + 1):
        result = num * i
        print(f"{num} × {i} = {result}")

    # Ask if user wants another table
    again = input("\nWould you like another table? (yes/no): ").lower()
    if again != "yes":
        print("Thank you for using the Multiplication Table Generator!")
        break
````
