## 9. Comparison Operators

Compare values using: `==` (equal), `!=` (not equal), `>` (greater), `<` (less), `>=` (greater or equal), `<=` (less or equal)

Comparisons return `True` or `False`

````python
# EXAMPLE: Comparisons
age = 20
limit = 18

print(age > limit)   # Is age greater than 18? True
print(age < limit)   # Is age less than 18? False
print(age == limit)  # Is age equal to 18? False
print(age >= limit)  # Is age greater or equal to 18? True
````

## 10. Logical Operators

Combine conditions using: `and`, `or`, `not`
- `and`: Both conditions must be True
- `or`: At least one condition must be True
- `not`: Reverses True/False

````python
# EXAMPLE: and operator
age = 20
has_license = True

can_drive = (age >= 16) and (has_license == True)
print("Can drive:", can_drive)  # True (both conditions are true)
````

## Solution Example

Here's one way to solve the student profile program. Try to solve it yourself first, then check this example.

**Key concepts to use:**
- Variables: `name`, `age`, `student_id`, `gpa`, `current_year`, `years_remaining`
- Arithmetic: Calculate `graduation_year = current_year + years_remaining`
- Comparison: Check `gpa >= 3.5` for honors status
- Logic: Check `(years_remaining > 1)` for advanced course eligibility
- Type checking: Use `type()` to verify variable types
- String concatenation: Combine strings to format output

````python
# SOLUTION EXAMPLE: Student Profile Program
name = "Alice Johnson"
age = 20
student_id = 98765
gpa = 3.8
current_year = 2026
years_remaining = 2

# Calculate graduation year (arithmetic)
graduation_year = current_year + years_remaining

# Check honors status (comparison)
is_honors_student = gpa >= 3.5

# Check course eligibility (logic + comparison)
can_register_advanced = years_remaining > 1

# Display results
print("=== STUDENT PROFILE ===")
print("Name:", name)
print("Age:", age)
print("Student ID:", student_id)
print("GPA:", gpa)
print("Expected Graduation Year:", graduation_year)
print("Is Honors Student:", is_honors_student)
print("Can Register for Advanced Courses:", can_register_advanced)

# Type verification
print("\n=== DATA TYPES ===")
print("Name type:", type(name))
print("Age type:", type(age))
print("GPA type:", type(gpa))
print("Is Honors Student type:", type(is_honors_student))
````
