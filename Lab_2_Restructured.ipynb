## 1. The if Statement - Making Decisions

**What is an if statement?**

An `if` statement lets your program make a decision. If a condition is `True`, the code inside runs. If `False`, it skips.

**Syntax:**
```python
if condition:
    code_block  # Runs only if condition is True
```

**How it works:**
1. Python evaluates the condition (is it True or False?)
2. If `True` â†’ executes the indented code
3. If `False` â†’ skips the code and continues with the rest of the program

## 2. The if-else Statement - Two Paths

**What is if-else?**

Sometimes you need two different actions based on a condition. The `if-else` statement provides two paths:
- If condition is `True` â†’ do one thing
- If condition is `False` â†’ do something different

**Syntax:**
```python
if condition:
    code_block_1  # Runs if True
else:
    code_block_2  # Runs if False
```

**When to use:** Use if-else when you have exactly two choices
