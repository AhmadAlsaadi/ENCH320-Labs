## Concept 1: if Statements - Making Decisions

An `if` statement allows your program to make decisions based on conditions.

**Syntax:**
```python
if condition:
    # Code executes if condition is True
```

**How it works:**
1. The condition is evaluated (True or False)
2. If True, the indented code block executes
3. If False, the code block is skipped

**Key Points:**
- Conditions evaluate to True or False
- Use comparison operators: `==`, `!=`, `<`, `>`, `<=`, `>=`
- Indentation (4 spaces) is critical in Python
- The colon `:` ends the condition line

## Concept 2: if-else Statements - Two Paths

An `if-else` statement provides two execution paths: one when the condition is True, another when it's False.

**Syntax:**
```python
if condition:
    # Code if condition is True
else:
    # Code if condition is False
```

**Key Points:**
- Exactly one code block executes
- Use `else` to handle the False case
- Perfect for binary (yes/no) decisions
- The program always takes one path

## Concept 3: if-elif-else Statements - Multiple Paths

When you need to check multiple conditions, use `if-elif-else`. The program checks each condition in order and executes the first True block.

**Syntax:**
```python
if condition1:
    # Code if condition1 is True
elif condition2:
    # Code if condition2 is True
elif condition3:
    # Code if condition3 is True
else:
    # Code if all conditions are False
```

**Key Points:**
- `elif` stands for "else if"
- Can have multiple `elif` blocks
- Only the first True condition executes
- `else` catches all remaining cases (optional)

## Concept 4: Nested Conditionals - Complex Decisions

Nested conditionals are `if` statements inside other `if` statements. Use them for complex, hierarchical decisions.

**Example:**
```python
if age >= 18:
    if has_license:
        print("You can drive!")
```

**Key Points:**
- Inner conditions only check if outer condition is True
- Use proper indentation (8 spaces for nested)
- Avoid deeply nested code (use `elif` or `and` operator instead)
- Can use with `and`, `or`, `not` operators
