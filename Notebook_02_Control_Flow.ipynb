## 1. The if Statement - Making Decisions

**What is an if statement?**

An `if` statement lets your program make a decision. If a condition is `True`, the code inside runs. If `False`, it skips.

**Syntax:**
```python
if condition:
    code_block  # Runs only if condition is True
```

**How it works:**
1. Python evaluates the condition (is it True or False?)
2. If `True` â†’ executes the indented code block
3. If `False` â†’ skips the code and continues with the rest of the program

**Examples of conditions:**
- `age >= 18` (age is 18 or greater)
- `score > 90` (score is greater than 90)
- `name == "Alice"` (name equals Alice)
- `is_student` (if the value is True)

- Else if third condition is `True` â†’ do something else
- Else (none are True) â†’ do the default

**Syntax:**
```python

    code_block_1

    code_block_2

    code_block_3
else:
    code_block_4
```

**Important:** Only ONE code block runs - the first condition that's True

## 6. range() Function - Three Formats

**What is range()?**

The `range()` function generates a sequence of numbers. It's most commonly used with for loops.

**Three formats:**

1. **`range(stop)`** - Counts from 0 up to (but not including) stop
   - `range(5)` produces: 0, 1, 2, 3, 4

   - `range(1, 6)` produces: 1, 2, 3, 4, 5

   - `range(0, 10, 2)` produces: 0, 2, 4, 6, 8
   - `range(5, 0, -1)` produces: 5, 4, 3, 2, 1

```

1. Python evaluates the condition (is it True or False?)
2. If `True` â†’ executes the indented code block
3. If `False` â†’ skips the code and continues with the rest of the program

- `age >= 18` (age is 18 or greater)

- `name == "Alice"` (name equals Alice)

## 6. range() Function - Three Formats

**What is range()?**

The `range()` function generates a sequence of numbers. It's most commonly used with for loops.

**Three formats:**

1. **`range(stop)`** - Counts from 0 up to (but not including) stop
   - `range(5)` produces: 0, 1, 2, 3, 4

   - `range(1, 6)` produces: 1, 2, 3, 4, 5

   - `range(0, 10, 2)` produces: 0, 2, 4, 6, 8
   - `range(5, 0, -1)` produces: 5, 4, 3, 2, 1

## 6. range() Function - Three Formats

**What is range()?**

The `range()` function generates a sequence of numbers. It's most commonly used with for loops.

**Three formats:**

1. **`range(stop)`** - Counts from 0 up to (but not including) stop
   - `range(5)` produces: 0, 1, 2, 3, 4

2. **`range(start, stop)`** - Counts from start up to (but not including) stop
   - `range(1, 6)` produces: 1, 2, 3, 4, 5

3. **`range(start, stop, step)`** - Counts from start to stop with a specific increment
   - `range(0, 10, 2)` produces: 0, 2, 4, 6, 8
   - `range(5, 0, -1)` produces: 5, 4, 3, 2, 1

## 10. The continue Statement - Skip to Next Iteration

**What is continue?**

The `continue` statement skips the rest of the current iteration and jumps to the next one.

**Syntax:**
```python
for item in items:
    if condition:
        continue  # Skip to next iteration
    code  # This is skipped if condition is True
```

**When to use:** When you want to skip certain items but continue with others
