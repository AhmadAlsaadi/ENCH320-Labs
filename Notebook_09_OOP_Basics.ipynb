## 1. Classes and Objects

### Creating a Simple Class

# Define a simple class
class Dog:
    # Constructor
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    # Method
    def bark(self):
        return f"{self.name} says Woof!"
    
    def get_age(self):
        return f"{self.name} is {self.age} years old"
    
    def __str__(self):
        return f"Dog(name={self.name}, age={self.age})"

# Create objects
dog1 = Dog("Buddy", 3)
dog2 = Dog("Max", 5)

print(dog1)
print(dog1.bark())
print(dog1.get_age())

print("\n" + str(dog2))
print(dog2.bark())

### Accessing Attributes

# Accessing and modifying attributes
dog = Dog("Charlie", 4)

print(f"Name: {dog.name}")
print(f"Age: {dog.age}")

# Modify attributes
dog.age = 5
print(f"Updated age: {dog.age}")

# Add new attribute
dog.color = "Brown"
print(f"Color: {dog.color}")

## 2. Class and Instance Variables

# Class variables vs instance variables
class Car:
    # Class variable (shared by all instances)
    total_cars = 0
    
    def __init__(self, brand, model):
        # Instance variables
        self.brand = brand
        self.model = model
        # Increment class variable
        Car.total_cars += 1
    
    def info(self):
        return f"{self.brand} {self.model}"
    
    @classmethod
    def get_total(cls):
        return f"Total cars: {cls.total_cars}"

car1 = Car("Toyota", "Camry")
car2 = Car("Honda", "Civic")
car3 = Car("Ford", "Mustang")

print(car1.info())
print(car2.info())
print(Car.get_total())

## 3. Inheritance

### Basic Inheritance

# Inheritance
class Animal:
    def __init__(self, name):
        self.name = name
    
    def speak(self):
        return f"{self.name} makes a sound"

# Inherit from Animal
class Cat(Animal):
    def speak(self):
        return f"{self.name} says Meow!"

class Dog(Animal):
    def speak(self):
        return f"{self.name} says Woof!"

# Create objects
cat = Cat("Whiskers")
dog = Dog("Rex")

print(cat.speak())
print(dog.speak())

# Check inheritance
print(f"\nIs cat an instance of Animal? {isinstance(cat, Animal)}")
print(f"Is cat an instance of Cat? {isinstance(cat, Cat)}")

### Using super()

# Using super() to call parent methods
class Vehicle:
    def __init__(self, brand, year):
        self.brand = brand
        self.year = year
    
    def info(self):
        return f"{self.year} {self.brand}"

class Truck(Vehicle):
    def __init__(self, brand, year, capacity):
        super().__init__(brand, year)  # Call parent constructor
        self.capacity = capacity
    
    def info(self):
        parent_info = super().info()  # Call parent method
        return f"{parent_info} (Capacity: {self.capacity} tons)"

truck = Truck("Volvo", 2022, 10)
print(truck.info())

## 4. Special Methods (__dunder__ methods)

# Special methods
class Book:
    def __init__(self, title, author, pages):
        self.title = title
        self.author = author
        self.pages = pages
    
    def __str__(self):
        return f"{self.title} by {self.author}"
    
    def __repr__(self):
        return f"Book('{self.title}', '{self.author}', {self.pages})"
    
    def __len__(self):
        return self.pages
    
    def __eq__(self, other):
        return self.pages == other.pages
    
    def __lt__(self, other):
        return self.pages < other.pages

book1 = Book("Python Basics", "John Doe", 300)
book2 = Book("Advanced Python", "Jane Smith", 450)

print(f"str: {str(book1)}")
print(f"repr: {repr(book1)}")
print(f"Length: {len(book1)}")
print(f"book1 == book2: {book1 == book2}")
print(f"book1 < book2: {book1 < book2}")

## 5. Properties and Getters/Setters

# Using properties
class Person:
    def __init__(self, name, age):
        self._name = name
        self._age = age
    
    @property
    def age(self):
        return self._age
    
    @age.setter
    def age(self, value):
        if value < 0:
            raise ValueError("Age cannot be negative")
        self._age = value
    
    def __str__(self):
        return f"{self._name} is {self._age} years old"

person = Person("Alice", 25)
print(person)

# Use property like an attribute
print(f"Age: {person.age}")

# Set using property
person.age = 26
print(person)

# person.age = -5
