### الدرس المصغر 5.1: تعديل متغير عام

## الدرس المصغر 6: التوثيق وإشارات النوع
**سلسلة التوثيق** تشرح ما تفعله الدالة. وإشارات النوع توضح نوع المدخلات والمخرجات.

**تحقق سريع:** ماذا تقول سلسلة التوثيق لـ `celsius_to_fahrenheit`؟

**جرّبها:** غيّر الإدخال إلى 25°C وتوقع الناتج.

</div>
</details>

```python
# Function with docstring and type hints
def celsius_to_fahrenheit(celsius: float) -> float:
    """
    Convert temperature from Celsius to Fahrenheit.

    Parameters:
    celsius (float): Temperature in Celsius

    Returns:
    float: Temperature in Fahrenheit
    """
    fahrenheit = (celsius * 9/5) + 32
    return fahrenheit

# Access the docstring
print(celsius_to_fahrenheit.__doc__)
print(f"0°C = {celsius_to_fahrenheit(0)}°F")
```

## الدرس المصغر 7: أفضل ممارسات الدوال
الدالة الجيدة تكون:
- قصيرة ومركزة
- باسم واضح
- بمدخلات ومخرجات واضحة

**تحقق سريع:** لماذا تُرجع `calculate_average` القيمة 0 عند قائمة فارغة؟

</div>
</details>

```python
# Good function example
from typing import List

def calculate_average(numbers: List[float]) -> float:
    """
    Calculate the average of a list of numbers.

    Args:
        numbers (list): A list of numeric values

    Returns:
        float: The average of the numbers
    """
    if len(numbers) == 0:
        return 0

    total = sum(numbers)
    average = total / len(numbers)
    return average

grades = [85, 90, 78, 92, 88]
avg = calculate_average(grades)
print(f"Average grade: {avg:.2f}")
```
